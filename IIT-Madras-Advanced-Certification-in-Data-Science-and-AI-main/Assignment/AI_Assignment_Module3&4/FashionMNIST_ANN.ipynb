```python
# loading dataset into train & test variables
fashion_mnist = keras.datasets.fashion_mnist

(train_images, train_labels) , (test_images, test_labels) = fashion_mnist.load_data()

# df = pd.read_csv('/content/fashion-mnist_train.csv')
# df_test = pd.read_csv('/content/fashion-mnist_test.csv')
```

```python
df = pd.DataFrame(train_images.reshape(60000,784))
df_test = pd.DataFrame(test_images.reshape(10000, 784))
```

```python
from keras.utils import np_utils
from keras.models import Sequential
from keras.layers import Dense, Activation
```

```python
x_train = df.values[:,:] / 255
y_train = train_labels
y_train = np_utils.to_categorical(train_labels)

x_test = df_test.values[:,:] / 255
y_test = test_labels
y_test = np_utils.to_categorical(y_test)

print("Training Dataset")
print(x_train.shape)
print(y_train.shape)

print("Testing Dataset")
print(x_test.shape)
print(y_test.shape)
```

```python
from tensorflow.python.ops.math_ops import dense_bincount
from keras.layers.reshaping.flatten import Flatten
# running this model for different batch_size with 20 epochs

def model_training(trainX, trainY, testX, testY, b_size):
  # defining model
    model = Sequential()

    model.add(Dense(512, input_shape = (784,)))
    model.add(Activation('relu'))

    model.add(Dense(256))
    model.add(Activation('relu'))

    model.add(Dense(10))
    model.add(Activation('softmax'))

  # compiling model
    model.compile(loss='categorical_crossentropy', optimizer = 'adam', metrics = ['accuracy'])

  # fit model
    history = model.fit(trainX, trainY, validation_data=(testX, testY), epochs=20, verbose=0, batch_size= b_size)

  # creating dataframe for (acc, val acc, loss, val loss) with different batch_size
    report = pd.DataFrame(history.history)
    return report
```
